```python
#(1)不使用函數與迴圈，直接計算五種T的買權價格
import numpy as np
from scipy.stats import norm

S = 90         # 標的資產價格
K = 100        # 履約價格
r = 0.01       # 無風險利率
sigma = 0.3    # 標的資產報酬率標準差

# T = 1
T = 1
d1 = (np.log(S/K) + (r + 0.5 * sigma**2)*T) / (sigma * np.sqrt(T))
d2 = d1 - sigma * np.sqrt(T)
call_1 = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)

# T = 0.9
T = 0.9
d1 = (np.log(S/K) + (r + 0.5 * sigma**2)*T) / (sigma * np.sqrt(T))
d2 = d1 - sigma * np.sqrt(T)
call_09 = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)

# T = 0.5
T = 0.5
d1 = (np.log(S/K) + (r + 0.5 * sigma**2)*T) / (sigma * np.sqrt(T))
d2 = d1 - sigma * np.sqrt(T)
call_05 = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)

# T = 0.3
T = 0.3
d1 = (np.log(S/K) + (r + 0.5 * sigma**2)*T) / (sigma * np.sqrt(T))
d2 = d1 - sigma * np.sqrt(T)
call_03 = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)

# T = 0.1
T = 0.1
d1 = (np.log(S/K) + (r + 0.5 * sigma**2)*T) / (sigma * np.sqrt(T))
d2 = d1 - sigma * np.sqrt(T)
call_01 = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)

# 輸出結果
print("T=1.0  Call =", round(call_1, 4))
print("T=0.9  Call =", round(call_09, 4))
print("T=0.5  Call =", round(call_05, 4))
print("T=0.3  Call =", round(call_03, 4))
print("T=0.1  Call =", round(call_01, 4))
```

```python
#(2)使用迴圈計算T = 1到 0.1歐式買權價格，並繪圖
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 解決中文字顯示問題
plt.rcParams['font.family'] = 'Microsoft JhengHei'
plt.rcParams['axes.unicode_minus'] = False

S = 90         # 標的資產價格
K = 100        # 履約價格
r = 0.01       # 無風險利率
sigma = 0.3    # 標的資產報酬率標準差

T_list = np.arange(1.0, 0.0, -0.1)
call_prices = []

for T in T_list:
    d1 = (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    call = S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    call_prices.append(call)

plt.figure(figsize=(8, 5))
plt.plot(T_list, call_prices, marker='o', linestyle='-', color='blue')
plt.title('歐式買權價格 vs 到期日 T')
plt.xlabel('到期日')
plt.ylabel('買權價格')
plt.grid(True)
plt.show()
```
